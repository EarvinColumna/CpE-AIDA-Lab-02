```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = [10, 15, 20, 25, 30, 35]
y = [68, 60, 55, 48, 38, 32]
```

```python
X_mean = np.mean(x)
Y_mean = np.mean(y)

n = len(x)
n = len(y)
```

```python
num = 0
den = 0
for i in range(n):
  num += (x[i] - X_mean) * (y[i] - Y_mean)
  den += (x[i] - X_mean) ** 2
A = num / den
F = Y_mean - (A * X_mean ) 
print('Coefficients:')
print(A, F)
```

```python
X_max = np.max(x) + 100
X_min = np.min(y) - 100

x = np.linspace(X_max, X_min, 1000)
y = F + A * x

plt.plot(x, y, color='black', label='Regression Line')
plt.scatter(x, y, c='orange', label='Scatter Plot')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
Error = 0
for i in range(n):
    Y_pred = F + A * x[i]
    Error += (y[i] - Y_pred) * 2
Error = np.sqrt(Error/n)
print("Error")
print(Error)
```

```python
W = 0
Z = 0
for i in range(n):
    Y_pred = F + A * x[i]
    W += (y[i] - Y_mean) * 2
    Z += (y[i] - Y_pred) * 2
R2 = 1 - (Z/W)
print('R2 Score:')
print(R2)
```

###**Question 6**

```python
def leastRegLine(X,Y):
	
  n = len(X)
  sxx = sum(np.multiply(X,X))
  sxy = sum(np.multiply(X,Y))


  b = (n * sxy - sum(X) * sum(Y))/(n * sxx - sum(X) * sum(X))
  meanX = sum(X)/n
  meanY = sum(Y)/n
 
  a = meanY - b * meanX

  print("Regression line:")
  print("Y = ", '%.3f'%a, " + ", '%.3f'%b, "*X", sep="")

  x = np.linspace(min(X),max(X),100)
  y = a + b*x

  plt.figure(figsize = (10,8))
  plt.plot(X,Y, "b.")
  plt.plot(x, y, 'r')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()
```

```python
def plotter(X,Y):
  return plt.plot(X,Y,"b.")
plotter([100,200,300,400,500,600],[228,236,256,278,285,301])
```
