```python
## 1 Is the no. of facebook likes in Director's Facebook page matters to the gross sales of the movie?
x = osx.director_facebook_likes
y = osx.gross
plt.scatter(x,y)
plt.xlabel ('Gross')
plt.ylabel ('Likes')

def find_cor(x,y):

  # Mean of x and y
  X = x.mean()
  Y = y.mean()

  # Difference for each data point from their means
  x_X = x - X
  y_Y = y - Y

  # Covariance of x and y
  cov_xy = np.cov(x,y, bias=True)[0,1]

  # Standard Deviation
  std_x = np.std(x)
  std_y = np.std(y)

  #Finding the correlation
  Corr = cov_xy / (std_x*std_y)
  return Corr


find_cor(x,y)
```

```python
#3. Does the plot of the movie dependent to budget? Show the positive and negative correlation if there is? 
x = osx.plot_keywords.str.count('\|') +1
x = x.replace(1,0).values
y = osx.budget
plt.scatter(x,y)
plt.xlabel('Plot Keywords')
plt.ylabel('Budget')

find_cor(x,y)
```
